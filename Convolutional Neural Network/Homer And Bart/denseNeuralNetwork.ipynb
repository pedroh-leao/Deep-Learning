# Using A Dense Neural Network To Classify Homer And Bart Simpson

```python
import tensorflow as tf
import pandas as pd
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import InputLayer, Dense, Dropout
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
dataset = pd.read_csv('characters.csv')
dataset
```

```python
x = dataset.iloc[:, 0:6].values
y = dataset.iloc[:, 6].values
```

```python
labelEncoder = LabelEncoder()
y = labelEncoder.fit_transform(y)
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.25)
```

```python
model = Sequential()
model.add(InputLayer(shape= (6, )))
model.add(Dense(units= 4, activation= 'relu'))
model.add(Dense(units= 4, activation= 'relu'))
model.add(Dense(units= 1, activation= 'sigmoid'))

model.compile(optimizer= tf.keras.optimizers.Adam(), loss= 'binary_crossentropy', metrics= ['binary_accuracy',])
```

```python
model.fit(
    x_train,
    y_train,
    epochs= 100,
    batch_size= 10,
    validation_data= (x_test, y_test)
)
```

```python
model.evaluate(x_test, y_test)
```
