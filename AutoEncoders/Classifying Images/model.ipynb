# AutoEncoder To Reduce The Dimensionality Of Image And Then Classify The Image

Comparing the results with the original and reduced dimensionality of the image.

Using mnist dataset.

```python
import tensorflow as tf
import keras
import numpy as np
```

```python
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.layers import Input, InputLayer, Dense
from tensorflow.keras import utils as np_utils
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()

x_train = x_train.astype('float32') / 255
x_test = x_test.astype('float32') / 255
```

```python
y_train = np_utils.to_categorical(y_train)
y_test = np_utils.to_categorical(y_test)
```

```python
x_train = x_train.reshape((len(x_train), np.prod(x_train.shape[1:])))
x_test = x_test.reshape((len(x_test), np.prod(x_test.shape[1:])))
```

```python
x_train.shape, x_test.shape
```

```python
model_ae = Sequential() # autoencoder

model_ae.add(InputLayer(shape= (784, )))
model_ae.add(Dense(units= 32, activation= 'relu'))
model_ae.add(Dense(units= 784, activation= 'sigmoid'))
```

```python
model_ae.compile(optimizer= tf.keras.optimizers.Adam(), loss= 'binary_crossentropy', metrics= ['accuracy'])
```

```python
model_ae.fit(
    x_train,
    x_train,
    epochs= 50,
    batch_size= 256,
    validation_data= (x_test, x_test)
)
```

```python
original_dimension = Input(shape= (784,))
encoder_layer = model_ae.layers[0]

encoder = Model(original_dimension, encoder_layer(original_dimension))
```

```python
encoded_training = encoder.predict(x_train)
```

```python
encoded_test = encoder.predict(x_test)
```

• Without Dimensionality Reduction:

```python
classifier1 = Sequential()
classifier1.add(InputLayer(shape= (784, )))
classifier1.add(Dense(units= 397, activation= 'relu'))
classifier1.add(Dense(units= 397, activation= 'relu'))
classifier1.add(Dense(units= 10, activation= 'softmax'))
```

```python
classifier1.compile(optimizer= tf.keras.optimizers.Adam(), loss= 'categorical_crossentropy', metrics= ['accuracy'])
```

```python
classifier1.fit(
    x_train,
    y_train,
    batch_size= 256,
    epochs= 100,
    validation_data= (x_test, y_test)
)
```

• With Dimensionality Reduction:

```python
classifier2 = Sequential()
classifier2.add(InputLayer(shape= (32, )))
classifier2.add(Dense(units= 21, activation= 'relu'))
classifier2.add(Dense(units= 21, activation= 'relu'))
classifier2.add(Dense(units= 10, activation= 'softmax'))
```

```python
classifier2.compile(optimizer= tf.keras.optimizers.Adam(), loss= 'categorical_crossentropy', metrics= ['accuracy'])
```

```python
classifier2.fit(
    encoded_training,
    y_train,
    batch_size= 256,
    epochs= 100,
    validation_data= (encoded_test, y_test)
)
```
