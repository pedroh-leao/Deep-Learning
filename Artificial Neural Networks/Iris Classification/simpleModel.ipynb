# Artificial Neural Network to classify the type of Iris flower
Using Iris dataset from Fisher

The dataset has three classes for the Iris: Iris Setosa, Iris Versicolour or Iris Virginica

```python
import tensorflow as tf
import pandas as pd
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras import utils as np_utils
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import InputLayer
from tensorflow.keras.models import Sequential
from sklearn.metrics import confusion_matrix
```

```python
dataset = pd.read_csv("iris.csv")
dataset
```

```python
# separating x and y
x = dataset.iloc[:, 0:4].values
x
```

```python
y = dataset.iloc[:, 4].values
y
```

```python
# encoding the classes
labelEncoder = LabelEncoder()
y = labelEncoder.fit_transform(y) # class -> num
y
```

```python
y = np_utils.to_categorical(y) # num of class -> one-hot representation
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.25)
```

```python
input_size = x.shape[1]
output_size = y.shape[1]

model = Sequential([
    InputLayer(shape= (input_size, )),
    Dense(units= round((input_size + output_size) / 2), activation= 'relu'),
    Dense(units= round((input_size + output_size) / 2), activation= 'relu'),
    Dense(units= output_size, activation= 'softmax')
])

model.summary()
```

```python
model.compile(optimizer= tf.keras.optimizers.Adam(learning_rate= 0.001), loss= 'categorical_crossentropy', metrics= ['categorical_accuracy', ])
```

```python
model.fit(
    x_train,
    y_train,
    epochs= 1000,
    batch_size= 10
)
```

```python
model.evaluate(x_test, y_test)
```

```python
predictions = model.predict(x_test)
predictions
```

```python
y_test
```

```python
predictions2 = [np.argmax(p) for p in predictions]
y_test2 = [np.argmax(p) for p in y_test]

print(predictions2)
print(y_test2)
```

```python
confusion_matrix(y_test2, predictions2)
```
